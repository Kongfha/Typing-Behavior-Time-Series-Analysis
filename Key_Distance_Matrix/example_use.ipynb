```python
import pandas as pd
import pickle

with open("adj_mat.pkl", "rb") as f:
    adj_mat = pickle.load(f)

with open("dist_mat.pkl", "rb") as f:
    dist_mat = pickle.load(f)

km = pd.concat([
    pd.read_csv("cap_eng_keycodes.csv"),
    pd.read_csv("small_eng_keycodes.csv"),
    pd.read_csv("thai_unshifted_keycodes.csv"),
    pd.read_csv("thai_shifted_keycodes.csv")
])
km = km.set_index("Key")["KeyCode"].to_dict()

print(adj_mat[km["ก"]][km["ง"]])
print(dist_mat[km["ก"]][km["ง"]])

print(adj_mat[km["A"]][km["ไ"]])
print(dist_mat[km["A"]][km["ง"]]) 
#Use dist_mat instead of adj_mat for distance, as adj_mat only indicates adjacency (1 or 0) while dist_mat provides the actual distance value.
```
